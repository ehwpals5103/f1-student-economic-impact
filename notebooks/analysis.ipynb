```python
# 데이터 정리: 실제 헤더는 row 1에 있음
students = pd.read_excel(
    "../data/Census_All-Places-of-Origin-Selected-Years_OD25_Website.xlsx",
    header=1
)

# 불필요한 행 제거 (NaN으로 된 빈 행들)
students = students.dropna(how='all')

# 컬럼 이름 확인
print("컬럼 목록:")
print(students.columns.tolist())
print("\n데이터 크기:", students.shape)
print("\n처음 10줄:")
students.head(10)
```

```python
# 헤더가 row 0에 있으니까 skiprows로 건너뛰기
students = pd.read_excel(
    "../data/Census_All-Places-of-Origin-Selected-Years_OD25_Website.xlsx",
    skiprows=2  # 처음 2줄 건너뛰기
)

# 컬럼 이름 확인
print("컬럼 목록:")
print(students.columns.tolist())
print("\n처음 5줄:")
students.head()
```

```python
# 불필요한 컬럼 제거 + 정리
students = students.drop(columns=['Unnamed: 0', 'Unnamed: 2'])

# NaN 행 제거 (빈 구분선 행들)
students = students.dropna(subset=['Place of Origin'])

# 대문자로만 된 행은 대륙/지역 카테고리 (AFRICA, SUB-SAHARAN 등)
# 개별 국가만 남기기 위해 대문자 행 제거
students_countries = students[students['Place of Origin'] != students['Place of Origin'].str.upper()].copy()

# '-' 를 0으로 변환
students_countries = students_countries.replace('-', 0)

# 최근 연도만 선택 (2000년대 이후)
year_cols = [c for c in students_countries.columns if '/' in str(c) and int(c.split('/')[0]) >= 2000]
recent = students_countries[['Place of Origin'] + year_cols].copy()

print("정리된 데이터:")
print(f"국가 수: {len(recent)}")
print(f"연도 범위: {year_cols[0]} ~ {year_cols[-1]}")
recent.head(10)
```

```python
# 미국 전체 유학생 수 트렌드 보기
# 연도별 총 유학생 수 합산
year_cols = [c for c in recent.columns if '/' in str(c)]
totals = recent[year_cols].apply(pd.to_numeric, errors='coerce').sum()

# 연도 라벨 정리 (2000/01 -> 2000)
years = [int(y.split('/')[0]) for y in totals.index]

# 시각화
plt.figure(figsize=(12, 6))
plt.plot(years, totals.values, marker='o', linewidth=2, color='#2563eb')
plt.title('Total International Students in the U.S. (2000-2024)', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Number of Students')
plt.grid(True, alpha=0.3)

# 코로나 표시
plt.axvline(x=2020, color='red', linestyle='--', alpha=0.5, label='COVID-19')
plt.legend()
plt.tight_layout()
plt.show()
```
